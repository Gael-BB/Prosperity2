```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('submission2_round_4.csv', header=0, sep=';')
data.head(10)
```

```python
products = ['AMETHYSTS', 'STARFRUIT', 'ORCHIDS', 'CHOCOLATE', 'STRAWBERRIES', 'ROSES', 'GIFT_BASKET', 'COCONUT', 'COCONUT_COUPON']
prices = {}
profits = {}
for product in products:
    prices[product] = data[data['product'] == product]['mid_price'].values
    profits[product] = data[data['product'] == product]['profit_and_loss'].values
    
```

```python
for product in products:
    fig, ax1 = plt.subplots()
    ax2 = ax1.twinx()
    ax1.plot(prices[product], label='price', color='red')
    ax2.plot(profits[product], label='profit', color='blue')
    ax1.set_ylabel('Price')
    ax2.set_ylabel('Profit')
    ax1.legend(loc='upper left')
    ax2.legend(loc='upper right')
    plt.title(product)
    plt.show()
```
