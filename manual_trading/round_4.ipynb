```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 100 
#calculate expected return for a given lower bid x , higher bid y
def exp_return(x, y, c):
    return ((x**2 / 100 * (1/100 - c) + c*x) * (100 - x)) + ((y**2 - x**2) / 100 * (1/100 - c) + c*(y - x)) * (100 - y)
```

```python
#find maximum return for all possible higher bids
optimal_returns = np.zeros(N+1)

for y in range(N+1):
    optimal_return = 0
    for x in range(y):
        z = exp_return(x,y,0)
        if z > optimal_return:
            optimal_return = z
    optimal_returns[y] = optimal_return
```

```python
#plot optimal returns against highest bids
highest_bids = np.linspace(0,100,101)
plt.plot(highest_bids,optimal_returns)
plt.xlabel("highest bid")
plt.ylabel("optimal return")
plt.show()
```

```python
#plot the drop in expected return past the optimal highest bid
profit_loss = optimal_returns[78:] - optimal_returns[78]
plt.plot(highest_bids[78:],profit_loss)
plt.xlabel("highest bid")
plt.ylabel("loss in profit")
plt.show()

```

```python
#calculate loss in profit if highest bid is below the average
avg_bid = 80
returns_below_avg = [optimal_returns[i] * (100-avg_bid)/(100-i) for i in range(avg_bid)]
profit_loss_below_avg = returns_below_avg - optimal_returns[78]

#plot loss in profit against highest bid
plt.plot(highest_bids[:80],profit_loss_below_avg)
plt.xlabel("highest bid")
plt.ylabel("loss in profit")
plt.show()
```

```python
#check optimal lower bid for the chosen higher bid
higher_bid = 85
expected_returns = [exp_return(x,higher_bid,0) for x in range(higher_bid)]
optimal_lower_bid = np.argmax(expected_returns)
print(optimal_lower_bid)
```
